#Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Creating Data frame

```python
dataframe= pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

#Convert the data type of column Rate

```python
def handlerate(value):
  value=str(value).split('/')
  value=value[0];
  return float(value)

dataframe['rate']=dataframe['rate'].apply(handlerate)
print(dataframe.head())

```

```python
dataframe.info()
```

#What type of restaurant do the majority of customers order from?

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel('Type of Resturent')
```

#Observations:
#The majority of the restaurent falls into the dinning category.

#Dinning restaurents are preffered by a larger number of individuals.

#How many votes has each type of restaurant received from customers?

```python
dataframe.head()
```

```python

# Group the DataFrame by 'listed_in(type)' and sum the 'votes' for each group
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()

# Create a new DataFrame with the grouped data
result = pd.DataFrame({'votes': grouped_data})

# Plot the results using a line plot
plt.plot(result, c="green", marker="o")

# Set the x-label and y-label with specified colors and font sizes
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
```

#Observations: The Dinning restuarent has received majority of the votes.

#Which type of restaurant received more offline orders, so that Zomato can provide those customers with some good offers?

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order',aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')

plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

#Observations: Dining restaurants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients prefer to place orders in person at restaurants, but prefer online ordering at cafes.
