#Convert the data type of column Rate

```python
def handlerate(value):
  value=str(value).split('/')
  value=value[0];
  return float(value)

dataframe['rate']=dataframe['rate'].apply(handlerate)
print(dataframe.head())

```

#How many votes has each type of restaurant received from customers?

```python
dataframe.head()
```

```python

# Group the DataFrame by 'listed_in(type)' and sum the 'votes' for each group
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()

# Create a new DataFrame with the grouped data
result = pd.DataFrame({'votes': grouped_data})

# Plot the results using a line plot
plt.plot(result, c="green", marker="o")

# Set the x-label and y-label with specified colors and font sizes
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
```
